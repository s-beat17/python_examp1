```python
import matplotlib.pyplot as plt

#plot 1:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = [0, 1, 2, 3]
y = [10, 20, 30, 40]

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
  
# First create some toy data:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]
  
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
  
fig.suptitle('matplotlib.pyplot.subplots() Example')
plt.show()
```

```python
import matplotlib.pyplot as plt
  
# First create some toy data:
x = [0, 1, 2, 3]
y = [3, 8, 1, 10]
  
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[1, 1].scatter(x, y)
  
fig.suptitle('Example')
plt.show()
```
